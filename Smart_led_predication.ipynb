In this model we used Deep neural network regresser to predict led light life cycle and send a notification to electric centre before 2 to 4 days to install new led light in that area.
LED light life cycle prediction depends on many parameters like weather(high temperature, rainy days etc), voltage fluctuation, high load, how old the LED is , how many times the  light gets on off .
 
So for better accuracy, we use third party data for prediction.
So in this model we used supervised learning.

Time dataset taken from ------>>> Santa Monica GIS Data Portal

# Now Start 


**Split Dataset into Train and Test**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, rat, test_size = 0.3, random_state = 101)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train, batch_size = 32, num_epochs = 1500, shuffle = True)
```

**In This we use DNNRegressor**

------>>>>
In this we use 6 hidden layer in NN

```python
dnn_model = tf.estimator.DNNRegressor(hidden_units = [20, 30, 16, 6, 10, 15], feature_columns = feat_cols)
```

```python
dnn_model.train(input_fn = input_func, steps = 25000)
```

```python
predict_input_func = tf.estimator.inputs.pandas_input_fn(
      x=X_test,
      batch_size=10,
      num_epochs=1,
      shuffle=False)
```

```python
pred_gen = dnn_model.predict(predict_input_func)
```

```python
predictions = list(pred_gen)
```

```python
final_preds = []
for pred in predictions:
    final_preds.append(pred['predictions'])
```
